```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,BatchNormalization,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
import numpy as np
np.unique(y_train)
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[0])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train.shape
```

```python
# we can flatten image by using numpy
X_train_flat=X_train.reshape(len(X_train),28*28)
X_test_flat=X_test.reshape(len(X_test),28*28)
```

```python
X_test_flat.shape
```

```python
X_train_flat.shape
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28))) # Flatten using keras
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=1000,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('Error')
plt.legend(['Train_Error','Test_Error'])
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.legend(['Train_Accuracy','Test_Accuracy'])
plt.show()
```

# Complete Deep Neural Network

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=10,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False
)
```

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))

model.add(Dense(1024, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(1024, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(512, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(256, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(256, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(256, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(256, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(128, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(64, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(32, activation="relu"))
model.add(Dropout(0.2))
model.add(BatchNormalization())

model.add(Dense(16, activation="relu"))
model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(X_test, y_test),batch_size=32, epochs=1000, callbacks=callback)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epochs')
plt.ylabel('Error')
plt.legend(['Train_Error','Test_Error'])
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epochs')
plt.ylabel('Accuracy')
plt.legend(['Train_Accuracy','Test_Accuracy'])
plt.show()
```
